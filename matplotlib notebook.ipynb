 ## Outlier
 - data point that are significant diffrrent from the rest of the dataset
 - extreme values either on right or left side in the distribution
 - outliers comes manually in the data
 ### outlier detection method
 - z-score
 - boxplot data is skewed
 - percentile based

## boxplot
- uses for outlier detection
- 5 number summary
- univariate analysis only for matplolib analysis
- multivariate analysis/bivariate:seaborn
- boxplot is applicable on numeric data
- it tells us the distribution

```python
df=sns.load_dataset("tips")
df.head()
```

```python
## detection of outlier in the tip column
plt.boxplot(df["tip"]); ## in this the distribution is the tip given more of 2 to 4
```

```python
df["tip"].describe()
```

```python
q1=df["tip"].quantile(0.25)
q1
```

```python
q3=df["tip"].quantile(0.75)
q3
```

```python
iqr=q3-q1
iqr
```

```python
upper_limit=q3+1.5*iqr
upper_limit
```

```python
lower_limit=q1-1.5*iqr
lower_limit
```

## Insight
- Any value of tip greater than 5.90625 and less than -0.34375 consider as my outlier in the data

```python
## now finding the outlier
df[df["tip"]>upper_limit]
```

```python
df[df["tip"]<lower_limit]
```

## Handling the outlier
- trimming
- capping

```python
## trimming
dff=df[(df["tip"]<=upper_limit)& (df["tip"]>=lower_limit)]
dff["tip"].skew()## to check the skewness
```

## capping


```python
new_df_cap = df.copy()  ## create a copy for handling outlier in tips
new_df_cap
```

```python
new_df_cap['tip'] = np.where(new_df_cap['tip'] > upper_limit, upper_limit, 
                                              np.where(new_df_cap['tip'] < lower_limit, lower_limit,
                                                       new_df_cap['tip'])) 
 
 
 
plt.boxplot(new_df_cap['tip']);
 
```
